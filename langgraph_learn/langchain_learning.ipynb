```python
from langchain_ollama.llms import OllamaLLM
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.messages import HumanMessage
from langchain_core.output_parsers import StrOutputParser

llm = OllamaLLM(
    model="gemma3:1b-it-qat",
    # model='gemma3:270m',
    temperature=0,
)
```

```python
joke_prompt = ChatPromptTemplate(
    [
        (
            "system",
            """
                You are the world class comedian
            """,
        ),
        (
            "human",
            """
                Tell me a joke with topic {joke_topic}      
            """,
        ),
    ]
)

joke_topic = "insect"

# response = llm.invoke([HumanMessage(query)])

chain = joke_prompt | llm
str_chain = chain | StrOutputParser()
joke_llm = str_chain.invoke({"joke_topic": joke_topic})
print(joke_llm)
```

```python
for chunk in str_chain.stream({"joke_topic": "beets"}):
    print(chunk, end="|")
```

```python
from datetime import date

current_date_prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            "you are the good assistant which answer the user question with indonesian language",
        ),
        (
            "human",
            """
            tanggal berapa sekarang? 
            current date: {date}
            """,
        ),
    ]
)

chain_human = current_date_prompt | llm

for chunk in chain_human.stream({"date": date.today()}):
    print(chunk, end="")
```

```python
SOURCE = """
Old Ship Saloon 2023 quarterly revenue numbers:
Q1: $174782.38
Q2: $467372.38
Q3: $474773.38
Q4: $389289.
"""

query_prompt = ChatPromptTemplate.from_messages([
    ("system", "you are the best data analyst in the world who provide user with the data"),
    ("human", """ {query_ship_shaloon}
     context: {SOURCE}
     """)
])

chain_rag = query_prompt | llm | StrOutputParser()

print(chain_rag)
for chunk in chain_rag.stream({"query_ship_shaloon":"what is the total revenue of old ship shaloon in Q2 and Q3 2023?", "SOURCE": SOURCE}):
    print(chunk, end="")
```

```python
### set debug

from langchain.globals import set_debug

set_debug(True)

from datetime import date

prompt = ChatPromptTemplate.from_messages([
    ("system", 'You know that the current date is "{current_date}".'),
    ("human", "{question}")
])

chain = prompt | llm | StrOutputParser()

chain.invoke({
    "question": "What is the current date?",
    "current_date": date.today()
})
```

```python
set_debug(False)

stream = chain.astream_events({
    "question": "What is the current date?",
    "current_date": date.today()
}, version="v1")

async for event in stream:
    print(event)
    print("-----")
```
